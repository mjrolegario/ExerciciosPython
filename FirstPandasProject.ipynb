```python
#importando o pandas
import pandas as pd 
#lendo o arquivo
tabela_work = pd.read_csv("Salarios.csv")
#mostrando a tabela de dados
display(tabela_work)
```

```python
#total de colunas e linhas da tabela 

linhas = tabela_work.shape[0]
colunas = tabela_work.shape[1]

print("A tabela tem o total de " + str(colunas) + " colunas e " + str(linhas) + " linhas.")
```

```python
#Qual Ã© o salÃ¡rio mÃ©dio?

coluna_salario = tabela_work[["TotalPay"]].mean()
display(coluna_salario)
```

```python
#Qual Ã© o salÃ¡rio mÃ¡ximo?

maior_salario = tabela_work["TotalPay"].max()

print(maior_salario)
```

```python
#Qual Ã© o emprego do JOSEPH DRISCOLL?

emprego_joseph = tabela_work.loc[tabela_work["EmployeeName"]=="JOSEPH DRISCOLL","JobTitle"]
display("O emprego de Joseph Driscoll Ã© " + emprego_joseph)
```

```python
#Qual o nome da pessoa que tem o maior salÃ¡rio de todos?

maior_salario = tabela_work["TotalPay"].max()
empregado_maiorSalario = tabela_work.loc[tabela_work["TotalPay"]==maior_salario,"EmployeeName"]

print(empregado_maiorSalario)
```

```python
#Qual Ã© o salÃ¡rio mÃ©dio por ano?

media_salarios = tabela_work[["Year", "TotalPay"]].groupby("Year").mean()

display(media_salarios)
```
